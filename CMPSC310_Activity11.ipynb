## Binomial Distribution

The numbers are randomly chosen from the random function. The graph is showing all of the randomly chosen variables of success. This is the result of doing 100 random trials with a 50% probability of getting a successful outcome.

```python
import numpy as np

n=100 # number of trials
p=0.5 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

This distribution is supposed to have a uniform type of distribution, where it is spread out evenly. This type of distribution seems like it would have more float values to be able to be spread evenly. None of the numbers appear to be the same.

```python
import numpy as np

uniform = np.random.uniform(low=1, high=10,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

This is running the experiment 100 times with no mean and 10% standard deviation. Most of the random variables go around the .1 of Std even though it can go up to 1. This shows that the random variable takes the sigma into account when making its random predictions.

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

## Bayes Algorithm

This algorithm is being used to train and test the iris data. The testing size is 50% of the data. Using the normal distribution of the data, there is less mislabeling in the prediction of the model.

Ref: https://scikit-learn.org/stable/modules/naive_bayes.html

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print(y_pred)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```
